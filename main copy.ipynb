```python
topwords50 = ['지역', '산림', '문화', '숲', '사람', '학습', '교육', '자연', '마을', '이용', '산', '나무', '관광', '생활', '역사', '마음', '활용', '이야기', '미래', '가을', '예술', '청소년', '전통', '삶', '음식', '여름', '소나무', '봄', '이해', '여성', '겨울', '지식', '공감', '정신', '공동체', '사상', '성인', '공예', '정서', '창조', '고급', '남성', '신앙', '교양', '유람', '유람기', '향유', '의식주', '포용', '금송', '성황당', '서낭당', '친화력']
```

```python
topwords_df = pd.DataFrame(columns=['명사', '빈도'])
for i in range(1,13):
    df = pd.read_csv('산문화/산+문화_2022_%s월_명사빈도.csv'%i)
    #df = df[df['명사'].isin(topwords50)]
    topwords_df = pd.merge(topwords_df, df, on='명사', how='outer', suffixes=('_df1', '_df2'))
    # NaN 값을 0으로 채우기
    topwords_df['빈도_df1'].fillna(0, inplace=True)
    topwords_df['빈도_df2'].fillna(0, inplace=True)

    # '빈도' column 값을 더하기
    topwords_df['빈도'] = topwords_df['빈도_df1'] + topwords_df['빈도_df2']
    # 필요한 column만 선택
    topwords_df = topwords_df[['명사', '빈도']]

topwords_df['빈도'] = topwords_df['빈도'].astype(int)
topwords_df = topwords_df.sort_values(by='빈도', ascending=False)
```

```python
#### 중복 단어 제거하여 리스트 생성
unique_lines = [list(set(line)) for line in tqdm(docs2)]
```

```python
#### 중복 제거 잘 됐나 개수로 확인
print(len(docs[0]))
print(len(unique_lines[0]))
print(unique_lines[0][0:10])
```

```python
for idx in tqdm(range(len(unique_lines))):
    unique_lines[idx] = list(set(topwords_df['명사']) & set(unique_lines[idx]))
```

```python
#### 중복 제거 잘 됐나 개수로 확인
print(len(docs[0]))
print(len(unique_lines[0]))
print(unique_lines[0][0:10])
```

```python
#### 전체 단어쌍 빈도 dict 형식으로 생성
freq_count = {}   #동시출현 빈도가 저장될 dict
for words in tqdm(unique_lines):
    # 전체 단어에 대한 동시 출현 빈도
    for i, a in enumerate(words):
        for b in words[i+1:]:
            if a == b: continue
            elif a>b:
                freq_count[b, a] = freq_count.get((b, a),0) + 1
            else :
                freq_count[a, b] = freq_count.get((a, b),0) + 1
```

```python
#### 딕셔너리를 데이터프레임에 담기
freq_df=pd.DataFrame.from_dict(freq_count, orient='index')
```

```python
#### dict형식을 컬럼명 지정해서 데이터프레임으로 만들기
list1 = [(freq_df.index[i][0], freq_df.index[i][1], freq_df[0][i]) for i in tqdm(range(len(freq_df)))]
freq_df2 = pd.DataFrame(list1, columns=["term1","term2","freq"])
print(freq_df2.shape)
freq_df2.head()
```

```python
#### 컬럼명 지정한 dict_to_데이터프레임 저장
freq_df2.to_excel("산문화/산문화_column_DICTtoDF.xlsx", index=False, encoding='UTF-8')
```

```python
#### 50위 이내에 속한 단어들로 이루어진 Dataframe을 추출
list2 =[(freq_df2.loc[i][0], freq_df2.loc[i][1], freq_df2.loc[i][2]) for i in tqdm(range(len(freq_df2))) if (freq_df2['term1'][i] in topwords50) &(freq_df2['term2'][i] in topwords50)]
freq_df3 = pd.DataFrame(list2, columns=["term1","term2","freq"])
freq_df3
```

```python
#### 상위50위 단어들로 이루어진 단어쌍 데이터프레임 저장
freq_df3.to_excel("산문화/산문화_top50_wordpair.xlsx", index=False, encoding='UTF-8')
```

```python
#### 공출현빈도(1-mode matrix) 만들기
word_co_matx = np.zeros((50,50))
mat_idx = topwords_df['명사'] # 단어동시출현행렬이니까 인덱스와 칼럼을 동일하게 지정
mat_col = topwords_df['명사']
word_co_matx = pd.DataFrame(word_co_matx, index=mat_idx, columns=mat_col)
print(word_co_matx.shape)
```

```python
#### 단어쌍의 첫 번째 단어가 인덱스와 같다 and 두 번째 단어가 칼럼과 같다 → 단어쌍빈도를 해당 셀 값에 삽입
for n in tqdm(range(len(freq_df3))):
    for idx in mat_idx:
        if idx == freq_df3['term1'][n]:
            for col in mat_col:
                if col == freq_df3['term2'][n]:
                    word_co_matx.loc[idx, col] = freq_df3['freq'][n]
                    word_co_matx.loc[col, idx] = freq_df3['freq'][n]
```

```python
#### 1-mode 매트릭스 확인하기
word_co_matx
```
